```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sum_series(N):
  j = np.arange(1, N+1)
  return (4 / (2 * j-1)) @ ((-1)** (j + 1))

def plot (N):
  values_s = [sum_series(n) for n in range(1, N+1)]
  Ns = np.arange(1, N+1)


#Plotting
  plt.plot(Ns, values_s, label="S(N)")
  plt.xlabel("N")
  plt.ylabel("S(N)")
  plt.title(" Convergence of S(N) to pi")
  plt.legend()
  plt.grid(True)
  plt.show()

plot(50)
```
